```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.arange(0,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('TITLE')
plt.show()
```

```python
plt.subplot(1,4,1)
plt.plot(y,x,'r')
plt.subplot(1,4,2)
plt.plot(x,y,'b')
plt.subplot(1,4,3)
plt.plot(y,x,'r')
plt.subplot(1,4,4)
plt.plot(x,y,'b')
```

## Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.9,0.4])
axes.plot(x,y)
```

```python
fig1 = plt.figure()

axes1 = fig1.add_axes([0.1,0.1,1,1])
axes2 = fig1.add_axes([0.2,0.2,0.6,0.6])
```

```python
fig2 = plt.figure()
axes = fig2.add_axes([0.1,0.1,0.5,1])  # 50% of the original width and 100% of the original height
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes1 = fig.add_axes([0.07,0.4,0.47,0.48])
axes.plot(x,y,'r')
axes1.plot(y,x,'b')
axes.set_title('LARGER PLOT')
axes1.set_title('SMALLER PLOT')
```

### SUBPLOTS USING OOP 

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'r')
axes[0].set_title('PLOT 1')
axes[1].plot(y,x)
axes[1].set_title('PLOT 2')

plt.tight_layout()
```

### FIG SIZE AND DPI

```python
fig = plt.figure(figsize=(13,2))
axe = fig.add_axes([0,0,1,1])
axe.plot(x,y)
```

```python
fig,axe = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axe[0].plot(x,y)
axe[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,lw=3,color='green',ls='--')
```
